# Scraping Jobs and internship
##### Sites: Internshala, Indeed, jobtimes.
- Using Beautifulsoup scraping all the data like company name, role, stipened, location from different site. 

```python
#Importing important libraries.
import requests
from bs4 import BeautifulSoup
```

# Internshala

```python
l = ["python", "data science", "front end development","web development", "data analytics"]

for i in l:
    for j in range(1,3):
        url = f"https://internshala.com/internships/keywords-{i}/page-{j}"
        r = requests.get(url).content
        soup = BeautifulSoup(r)

        coms = soup.find_all("div", class_="internship_meta")    
        for com in coms:

            try:
                company_name = com.find("div", class_="heading_6 company_name").text.replace("\n", "")
            except Exception as e:
                print("None")
            try:
                role = com.find("div", class_="heading_4_5 profile").text.replace("\n","")
            except Exception as e:
                print("None")
            try:
                location = com.find("a", class_="location_link view_detail_button").text
            except Exception as e:
                print("None")
            try:
                stipend = com.find("span", class_ ="stipend").text.replace(" /month", "")
            except Exception as e:
                print("None")
            try:
                info = com.find("a")["href"].replace(" ","")
            except Exception as e:
                print("None")

            print("Company Name: ",company_name)
            print("Role: ",role)
            print("Location: ",location)
            print("stiened: ",stipend)
            print(f"more info: https://internshala.com/{info}")
            print("-"*100)
    print("="*100)
```

# TimeJobs

```python
l = ["python", "data science", "Front+End+Developer", "web+develpoment", "data analyst"]

for i in l:
    url = f"https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords={i}&txtLocation=#"
    r = requests.get(url).content
    soup = BeautifulSoup(r)

    coms = soup.find_all("li", class_ ="clearfix job-bx wht-shd-bx")    
    for com in coms:

        try:
            company_name = com.find("h3", class_ ="joblist-comp-name").text
        except Exception as e:
            print("None")
        try:
#                 role = com.find("strong", class_ ="blkclor").text
            role = com.find("h2").text
        except Exception as e:
            print("None")
        try:
            skills = com.find("span", class_ ="srp-skills").text.replace(" \n", "")
        except Exception as e:
            print("None")
        try:
            info = com.find("a")["href"]
        except Exception as e:
            print("None")

        print(f"company Name: {company_name}")
        print(f"Role: {role}")
        print(f"Skills: {skills}")
        print(f"More Info: {info}")
        print("-"*100)
    print("="*100)
```

# Indeed

```python
l = ["python", "data science", "front end development","web development", "data analytics"]

for i in l:
    for j in range(0,30,10):

        url = f"https://in.indeed.com/jobs?q={i}&start={j}&vjk=54ae5b809131fd35"
        r = requests.get(url).content
        soup = BeautifulSoup(r)

        coms = soup.find_all("div", class_="job_seen_beacon")     
        for com in coms:
            try:
                role = com.find("h2", class_="jobTitle").text
            except Exception as e:
                print("None")
            try:
                info = com.find("a")["href"]
            except Exception as e:
                print("None")
            try:
                company_name = com.find("span", class_="companyName").text
            except Exception as e:
                print("None")
            try:
                location = com.find("div", class_="companyLocation").text
            except Exception as e:
                print("None")
            try:
                stipend = com.find("div", class_="attribute_snippet").text
            except Exception as e:
                print("None")

            print(f"Comapny Name: {company_name}")
            print(f"role: {role}")
            print(f"Location: {location}")
            print(f"stipend: {stipend}")
            print(f"https://in.indeed.com{info}")
            print("-"*100)
        print("="*100)
```
